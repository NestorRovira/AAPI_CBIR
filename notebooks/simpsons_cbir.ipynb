```python
import matplotlib.pyplot as plt
import cv2
import numpy as np 
import os
```

## Eliminación de duplicados

```python
def obtener_rutas_directorios(directorio):
    rutas = []
    for raiz,dirs, archivos in os.walk(directorio):
        for dir in dirs:
            rutas.append(os.path.join(raiz, dir))
    return rutas

rutas_dirs = obtener_rutas_directorios("the_simpson_dataset\\train")
print(rutas_dirs)
```

```python
def obtener_rutas_archivos(directorio):
    rutas = []
    for raiz, _, archivos in os.walk(directorio):
        for archivo in archivos:
            rutas.append(os.path.join(raiz, archivo))
    return rutas

rutas = obtener_rutas_archivos("the_simpson_dataset\\train\\marge_simpson")

```

```python
def eliminar_duplicados(rutas: list): 
    removed = []
    for i in range(0, len(rutas)-1): 
        if i not in removed:
            img1 = cv2.imread(rutas[i])
            img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

            hsv1 = cv2.cvtColor(img1, cv2.COLOR_RGB2HSV)

            hist1 = cv2.calcHist([hsv1], [0, 1], None, [50, 60], [0, 180, 0, 256])

            cv2.normalize(hist1, hist1)

        for j in range (0, len(rutas) - 1): 
            if i < j and j not in removed: 
                img2 = cv2.imread(rutas[j])
                img2= cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
             
                hsv2 = cv2.cvtColor(img2, cv2.COLOR_RGB2HSV)
                
                hist2 = cv2.calcHist([hsv2], [0, 1], None, [50, 60], [0, 180, 0, 256])

                cv2.normalize(hist2, hist2)

                score = cv2.compareHist(hist1, hist2, cv2.HISTCMP_CORREL)
                # print(score, i, j )
                
                if score > 0.85:   
                    os.remove(rutas[j])
                    removed += [j]
    # print ("Archivos eliminados en este directorio:", len(removed))
    return len(removed)             
```

```python
batch_size = 50
total_images = []
for ruta_dir in rutas_dirs:
    n_images = 0; i = 0
    rutas = obtener_rutas_archivos(ruta_dir) 
    if len(rutas) > 150:
        print("Recopilando en ", ruta_dir[26:]) 
        while n_images < 110 and i < 20:
            n_images += min(batch_size, len(rutas)- (i+1)*batch_size) - eliminar_duplicados(rutas[i*batch_size : min( (i+1)*batch_size, len(rutas) )], )
            i += 1
        print(n_images, "añadidas para", ruta_dir[26:])
    else: 
        print(ruta_dir[26:], "no es elegible")
    total_images += [n_images]
```
